## Check unique genes in tumor from 07 analysis in leto (before crush) 

```python
normal = pd.read_csv("/data/scratch/kvalem/projects/2023/tmp/NSCLC_gender_2023/04-tables/nsclc_gender_normal_all_genes_DESeq2_result.csv")
tumor = pd.read_csv("/data/scratch/kvalem/projects/2023/tmp/NSCLC_gender_2023/04-tables/nsclc_gender_tumor_all_genes_DESeq2_result.csv")
```

```python
normal
```

```python
# Find unique gene names in each dataframe
unique_to_normal = normal[~normal['gene_name'].isin(tumor['gene_name'])]
unique_to_tumor = tumor[~tumor['gene_name'].isin(normal['gene_name'])]

# Combine the unique rows
unique_rows = pd.concat([unique_to_normal, unique_to_tumor], ignore_index=True)
```

```python
unique_rows.to_csv("../04-tables/unique_to_tumor_DEG.csv")
```
